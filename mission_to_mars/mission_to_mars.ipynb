```python
# Dependencies 
from bs4 import BeautifulSoup
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# URL for the page to be scraped
url="https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
```

```python
# Retrive page with the requests module
response=requests.get(url)
```

```python
# Create BeautifulSoup object parse with lxml
soup=BeautifulSoup(response.text,"lxml")
```

```python
print(soup.prettify())
```

```python
news_title=soup.find("div",class_="content_title").text.strip()

news_title
```

```python
news_p = soup.find("div",class_="rollover_description_inner").text.strip()
news_p
```

```python
executable_path={"executable_path":ChromeDriverManager().install()}
browser=Browser("chrome",**executable_path,headless=False)
```

```python
url="https://www.jpl.nasa.gov/images/new-all-in-one-antenna-for-the-deep-space-network/"
browser.visit(url)
```

```python
# html object
html=browser.html
```

```python
#Parse HTML with Beautiful Soup
soup=BeautifulSoup(html,"html.parser")
print(soup.prettify())
```

```python
img_page = soup.find_all('div', class_='lg:w-auto w-full')
for img in img_page:
    img_place = img.find('span', class_='label block')
    print(img_place)
    for place in img_place:
        browser.find_link_by_text('place').first.click()

```
